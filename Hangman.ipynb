```python
print("Welcome to the game Hangman")
print("I am thinking of a word that is 5 letters long")
print("----------")
word="apple"
letters="abcdefghijklmnopqrstuvwxyz"
list_l = list(letters)
list_w= list(word)
list_g=[]
guesses_left = 8;
flag=0;
s="_____"
spaces=list(s)
while(guesses_left>0):
    print("You have ",guesses_left," guesses left.")
    print("Available Letters: ", *list_l)
    l=input("Please guess a letter: ")
    if(l in list_g):
        print("You have already tried this letter")
        print("----------\n\n\n")
        continue
    list_g.append(l)
    for x in range(5):
        if(list_w[x]==l):
            spaces[x]=l
            flag=1
    list_l.remove(l)    
    if(flag==1):
        flag=0
        print("Good guess: ", *spaces)
        if(spaces==list_w):
            print("Congratulations, you won!")
            guesses_left=-1
    else:
        print("Uh oh! Wrong guess")
        guesses_left-=1
        flag=0
    print("----------\n\n\n")
if(guesses_left==0):
    print("You ran out of guesses. Better luck next time!")
```
